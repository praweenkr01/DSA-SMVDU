```
from numpy import mean
from numpy import std
from numpy import cov
from numpy.random import randn
from numpy.random import seed
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import seaborn as sns
import pandas as pd
import numpy as np
from pandas.plotting import scatter_matrix
%matplotlib inline
```

```
data1=20* randn(1000)+100
data2= data1+(20*randn(1000)+50)
```

```
print('data1 : mean=%.3f std=%.3f' %(mean(data1), std(data1)))
```

```
print('data1 : mean=%.3f std=%.3f' %(mean(data2), std(data2)))
```

```
plt.scatter(data1,data2)
```

Covariance(x,y)=(1/(n-1))*(sum((x-mean(x))*(y-mean(y))))

```
covariance=cov(data1,data2)
```

```
covariance
```

Pearson's Correlation coefficient= covariance(X,Y)/ (stdv(Y)*std(Y))

```
corr,_=pearsonr(data1,data2)
```

```
corr
```
