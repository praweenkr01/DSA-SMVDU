```
import pandas as pd
from matplotlib import dates
```

```
df = pd.read_csv('Electric_Production.csv',index_col='DATE',parse_dates=True)
#data source : https://www.kaggle.com/shenba/time-series-datasets?select=Electric_Production.csv
```

```
df['IPG2211A2N'].plot();
```

```
title='Electric Production'
ylabel='IPG2211A2N'
xlabel='Date'

ax = df['IPG2211A2N'].plot(figsize=(12,6),title=title)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel);
```

```
df['IPG2211A2N']['2004-01-01':'2005-03-01'].plot(figsize=(12,4)).autoscale(axis='x',tight=True);
```

```
df['IPG2211A2N'].plot(figsize=(12,4),xlim=['2004-01-01','2005-03-01']);
```
